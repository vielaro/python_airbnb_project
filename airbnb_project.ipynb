```python
#df.dtypes
df['id'] = df['id'].astype(object)
df.dtypes

df['host_id'] = df['host_id'].astype(object)
df.dtypes
```

```python
df2 = df[df['price'] < 1500]

sns.boxplot(data=df2, x='price')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df2, x='price',bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df2.dtypes
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=df2, x='availability_365')
plt.title('availability_365')
plt.ylabel("Frequency")
plt.show()
```

```python
df2.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df2['price_per_bed'] = df2['price']/df2['beds']
df2.head()
```

```python
sns.barplot(data=df2, x= 'neighbourhood_group', y = 'price', hue='room_type')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Locality and Review Analysis')
sns.scatterplot(data=df2, x='price', y='number_of_reviews', hue='neighbourhood_group')
plt.show()
```

```python
df2.dtypes
```

```python
sns.pairplot(data=df2, vars=['price','minimum_nights','number_of_reviews', 'availability_365'], hue='room_type')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df2, x='longitude', y='latitude', hue='room_type')
plt.title('Geographical Distribution of Airbnb Listings')
plt.show()
```

```python
corr = df2[['latitude','longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

sns.heatmap(data=corr, annot=True)
plt.figure(figsize=(8,6))
plt.show()
```
